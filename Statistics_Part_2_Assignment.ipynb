#Statistics Part 2

# Q1: What is hypothesis testing in statistics
# A:
# Hypothesis testing is a method for making decisions or inferences about population parameters using sample data, by testing assumed statements (hypotheses).

# Q2: What is the null hypothesis, and how does it differ from the alternative hypothesis
# A:
# The null hypothesis (H0) claims no effect or difference exists. The alternative hypothesis (H1) claims there is an effect or difference to be detected.

# Q3: What is the significance level in hypothesis testing, and why is it important
# A:
# The significance level (alpha) is the probability threshold for rejecting H0, often 0.05. It controls the risk of Type 1 error (false positives).

# Q4: What does a P-value represent in hypothesis testing
# A:
# The P-value is the probability of observing the test statistic or something more extreme, assuming the null hypothesis is true.

# Q5: How do you interpret the P-value in hypothesis testing
# A:
# A small P-value (below alpha) suggests evidence against H0, leading to rejection of H0. A large P-value suggests insufficient evidence to reject H0.

# Q6: What are Type 1 and Type 2 errors in hypothesis testing
# A:
# Type 1 error: rejecting a true H0 (false positive). Type 2 error: failing to reject a false H0 (false negative).

# Q7: What is the difference between a one-tailed and a two-tailed test in hypothesis testing
# A:
# One-tailed tests look for an effect in one direction; two-tailed tests check for effects in both directions.

# Q8: What is the Z-test, and when is it used in hypothesis testing
# A:
# A Z-test evaluates whether a sample mean significantly differs from a population mean when population variance is known and sample size is large.

# Q9: How do you calculate the Z-score, and what does it represent in hypothesis testing
# A:
# Z = (X̄ - μ) / (σ/√n). It represents how many standard errors the sample mean is from the population mean under H0.

# Q10: What is the T-distribution, and when should it be used instead of the normal distribution
# A:
# The T-distribution accounts for extra uncertainty with small samples (n < 30) when population standard deviation is unknown.

# Q11: What is the difference between a Z-test and a T-test
# A:
# Z-test uses known population variance and large samples. T-test uses sample variance for small samples where population variance is unknown.

# Q12: What is the T-test, and how is it used in hypothesis testing
# A:
# A T-test compares sample means (or a sample mean to a known value) when variance is estimated from data, useful for small samples.

# Q13: What is the relationship between Z-test and T-test in hypothesis testing
# A:
# Both test means, but T-test adjusts for sample size and estimated variance; as sample size grows, the T-distribution approaches the normal distribution.

# Q14: What is a confidence interval, and how is it used to interpret statistical results
# A:
# A confidence interval gives a range of values likely to contain the true parameter, expressing estimation uncertainty.

# Q15: What is the margin of error, and how does it affect the confidence interval
# A:
# The margin of error is the range above/below the point estimate. Larger margins mean wider intervals, reflecting more uncertainty.

# Q16: How is Bayes' Theorem used in statistics, and what is its significance
# A:
# Bayes' Theorem updates the probability of a hypothesis given new evidence. It's key in Bayesian inference for revising beliefs.

# Q17: What is the Chi-square distribution, and when is it used
# A:
# A right-skewed distribution used for testing variance, independence in contingency tables, and goodness-of-fit for categorical data.

# Q18: What is the Chi-square goodness of fit test, and how is it applied
# A:
# It tests whether observed categorical frequencies match expected frequencies under a hypothesized distribution.

# Q19: What is the F-distribution, and when is it used in hypothesis testing
# A:
# A right-skewed distribution used for comparing variances or in ANOVA to test differences among multiple means.

# Q20: What is an ANOVA test, and what are its assumptions
# A:
# ANOVA tests for differences among group means. Assumptions: independence, normality, and equal variances.

# Q21: What are the different types of ANOVA tests
# A:
# One-way ANOVA (one factor), two-way ANOVA (two factors), repeated measures ANOVA (same subjects multiple times).

# Q22: What is the F-test, and how does it relate to hypothesis testing
# A:
# The F-test compares variances to determine if groups have significantly different variability, used in ANOVA and regression.

```python
# Q1: Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
# A:
import numpy as np
from scipy.stats import norm

sample = np.random.normal(50, 10, 30)
pop_mean = 52
pop_std = 10

sample_mean = np.mean(sample)
se = pop_std / np.sqrt(len(sample))
z = (sample_mean - pop_mean) / se
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z:.2f}")
print(f"P-value: {p_value:.4f}")
```

```python
# Q4: Perform a two-tailed Z-test using Python and visualize the decision region on a plot
# A:
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

sample = np.random.normal(100, 15, 50)
pop_mean = 105
pop_std = 15

sample_mean = np.mean(sample)
se = pop_std / np.sqrt(len(sample))
z = (sample_mean - pop_mean) / se
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"Z-statistic: {z:.2f}")
print(f"P-value: {p_value:.4f}")

x = np.linspace(-4, 4, 1000)
plt.plot(x, norm.pdf(x))
plt.fill_between(x, 0, norm.pdf(x), where=(x < -1.96) | (x > 1.96), color='red', alpha=0.3, label='Rejection Region (5%)')
plt.axvline(z, color='black', linestyle='--', label=f'Z = {z:.2f}')
plt.title('Two-tailed Z-test Decision Regions')
plt.legend()
plt.show()
```

```python
# Q5: Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
# A:
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2(mu0, mu1, sigma, alpha=0.05):
    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)
    z_alpha = norm.ppf(1 - alpha/2)

    plt.plot(x, norm.pdf(x, mu0, sigma), label='Null (H0)')
    plt.plot(x, norm.pdf(x, mu1, sigma), label='Alternative (H1)')

    crit_low = mu0 - z_alpha * sigma
    crit_high = mu0 + z_alpha * sigma
    plt.axvline(crit_low, color='red', linestyle='--', label='Critical Region')
    plt.axvline(crit_high, color='red', linestyle='--')

    plt.fill_between(x, 0, norm.pdf(x, mu0, sigma), where=(x < crit_low) | (x > crit_high), color='red', alpha=0.3, label='Type 1 Error')
    plt.fill_between(x, 0, norm.pdf(x, mu1, sigma), where=(crit_low <= x) & (x <= crit_high), color='blue', alpha=0.3, label='Type 2 Error')

    plt.legend()
    plt.title('Type 1 and Type 2 Errors Visualization')
    plt.show()

plot_type1_type2(0, 1.5, 1)
```

```python
# Q6: Write a Python program to perform an independent T-test and interpret the results
# A:
import numpy as np
from scipy.stats import ttest_ind

group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 10, 30)

t_stat, p_value = ttest_ind(group1, group2)

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Significant difference between groups.")
else:
    print("No significant difference between groups.")
```

```python
# Q8: Simulate data and perform both Z-test and T-test, then compare the results using Python
# A:
import numpy as np
from scipy.stats import norm, ttest_1samp

data = np.random.normal(100, 15, 30)
pop_mean = 105
pop_std = 15

# Z-test
sample_mean = np.mean(data)
se_z = pop_std / np.sqrt(len(data))
z = (sample_mean - pop_mean) / se_z
p_z = 2 * (1 - norm.cdf(abs(z)))

# T-test
t_stat, p_t = ttest_1samp(data, pop_mean)

print(f"Z-test: Z = {z:.2f}, P-value = {p_z:.4f}")
print(f"T-test: T = {t_stat:.2f}, P-value = {p_t:.4f}")
```

```python
# Q9: Write a Python function to calculate the confidence interval for a sample mean and explain its significance
# A:
import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    se = np.std(data, ddof=1) / np.sqrt(len(data))
    z = norm.ppf(1 - (1 - confidence) / 2)
    lower = mean - z * se
    upper = mean + z * se
    return mean, lower, upper

data = np.random.normal(50, 8, 40)
mean, lower, upper = confidence_interval(data)

print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
# Q10: Write a Python program to calculate the margin of error for a given confidence level using sample data
# A:
import numpy as np
from scipy.stats import norm

data = np.random.normal(100, 15, 50)
confidence = 0.95
se = np.std(data, ddof=1) / np.sqrt(len(data))
z = norm.ppf(1 - (1 - confidence) / 2)
margin_error = z * se

print(f"Margin of Error (95% confidence): ±{margin_error:.2f}")
```

```python
# Q11: Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
# A:
def bayes_theorem(prior, likelihood, evidence):
    posterior = (likelihood * prior) / evidence
    return posterior

prior = 0.01
likelihood = 0.95
false_alarm = 0.05
evidence = likelihood * prior + false_alarm * (1 - prior)

posterior = bayes_theorem(prior, likelihood, evidence)

print(f"Posterior Probability: {posterior:.4f}")
```

```python
# Q16: Implement an F-test using Python to compare the variances of two random samples
# A:
import numpy as np
from scipy.stats import f

data1 = np.random.normal(50, 10, 30)
data2 = np.random.normal(50, 15, 30)

var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)
F = var1 / var2 if var1 > var2 else var2 / var1
dof1 = len(data1) - 1
dof2 = len(data2) - 1

p_value = 1 - f.cdf(F, dof1, dof2)

print(f"F-statistic: {F:.2f}")
print(f"P-value: {p_value:.4f}")
```

```python
# Q19: Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
# A:
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups):
    all_data = np.concatenate(groups)
    print("Normality (Shapiro-Wilk):")
    for i, group in enumerate(groups, 1):
        stat, p = shapiro(group)
        print(f"  Group {i}: p = {p:.4f}")
    print("\nHomogeneity of Variances (Levene's Test):")
    stat, p = levene(*groups)
    print(f"  Levene's test p-value = {p:.4f}")
    print("\nVisualizing:")
    sns.boxplot(data=[*groups])
    plt.title('Boxplots for Visual Inspection')
    plt.show()

# Example
group1 = np.random.normal(50, 5, 30)
group2 = np.random.normal(52, 5, 30)
group3 = np.random.normal(54, 5, 30)
check_anova_assumptions(group1, group2, group3)
```

```python
# Q20: Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results
# A:
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns

# Simulate data
np.random.seed(0)
factorA = np.repeat(['Low', 'High'], 30)
factorB = np.tile(np.repeat(['Type1', 'Type2', 'Type3'], 10), 2)
values = np.random.normal(50, 5, 60) + (factorA == 'High') * 5 + (factorB == 'Type2') * 3

data = pd.DataFrame({'FactorA': factorA, 'FactorB': factorB, 'Value': values})

# Two-way ANOVA
model = ols('Value ~ FactorA * FactorB', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)

# Visualization
sns.boxplot(x='FactorA', y='Value', hue='FactorB', data=data)
plt.title('Two-way ANOVA: Interaction Effect')
plt.show()
```

```python
# Q24: Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
# A:
import numpy as np
from scipy.stats import chi2

data = np.random.normal(100, 12, 30)
sample_var = np.var(data, ddof=1)
n = len(data)
hyp_var = 10**2

chi2_stat = (n - 1) * sample_var / hyp_var
p_value = 1 - chi2.cdf(chi2_stat, n - 1)

print(f"Chi-square Statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.4f}")
```

```python
# Q26: Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results
# A:
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

data1 = np.random.normal(50, 8, 30)
data2 = np.random.normal(50, 12, 30)

var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)
F_stat = var1 / var2 if var1 > var2 else var2 / var1
dof1, dof2 = len(data1) - 1, len(data2) - 1
p_value = 1 - f.cdf(F_stat, dof1, dof2)

print(f"F-statistic: {F_stat:.2f}")
print(f"P-value: {p_value:.4f}")

plt.boxplot([data1, data2], labels=['Data1', 'Data2'])
plt.title('Variance Comparison')
plt.show()
```
